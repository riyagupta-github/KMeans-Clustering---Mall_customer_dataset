```python
# Import necessary libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import os
import warnings

# Ignore warnings
warnings.filterwarnings('ignore')
```

```python
# Read the CSV file into a pandas DataFrame
data = pd.read_csv('Mall_Customers.csv')
data
```

```python
# Selecting values from the DataFrame 'data' using iloc to extract all rows and columns 3 to 4 (exclusive) and storing it in variable 'x'
x = data.iloc[:,3:5].values
x
```

```python
# Importing the KMeans module from the sklearn.cluster library
from sklearn.cluster import KMeans

abs# Creating an empty list to store the within-cluster sum of squares (WCSS) for different numbers of clusters
wcss = []

# Looping through a range of numbers from 2 to 10
for i in range(2, 11):
    # Creating a KMeans model with the current number of clusters, using k-means++ initialization and 10 iterations
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10)
    
    # Fitting the KMeans model to the data 'x'
    kmeans.fit(x)
    
    # Appending the WCSS of the current KMeans model to the list
    wcss.append(kmeans.inertia_)

# Displaying the list of WCSS values for different numbers of clusters
display(wcss)
```

```python
# Create a figure with a specific size
plt.figure(figsize=(20, 10))

# Plot the range of clusters against the within-cluster sum of squares (wcss)
plt.plot(range(2, 11), wcss)

# Set the title of the plot
plt.title('Elbow Method')

# Set the label for the x-axis
plt.xlabel('Number of clusters')

# Set the label for the y-axis
plt.ylabel('wcss')

# Display the plot
plt.show()
```

```python
kmeans = KMeans(n_clusters= 5, init= 'k-means++')
kmeans = kmeans.fit(x)
y_kmeans = kmeans.labels_
display(y_kmeans)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter (x[y_kmeans == 0,0],x[y_kmeans == 0,1] ,s = 100, c='r', label = 'Cluster 1')
plt.scatter (x[y_kmeans == 1,0],x[y_kmeans == 1,1] ,s = 100, c='b', label = 'Cluster 2')
plt.scatter (x[y_kmeans == 2,0],x[y_kmeans == 2,1] ,s = 100, c='g', label = 'Cluster 3')
plt.scatter (x[y_kmeans == 3,0],x[y_kmeans == 3,1] ,s = 100, c='c', label = 'Cluster 4')
plt.scatter (x[y_kmeans == 4,0],x[y_kmeans == 4,1] ,s = 100, c='m', label = 'Cluster 5')
plt.legend()
plt.show()
```

```python
# Initialize KMeans with 8 clusters using k-means++ initialization
kmeans = KMeans(n_clusters=8, init='k-means++')

# Fit the model to the data
kmeans = kmeans.fit(x)

# Get the cluster labels
y_kmeans = kmeans.labels_

# Display the cluster labels
display(y_kmeans)

# Create a scatter plot with different colors for each cluster
plt.figure(figsize=(20, 10))
for i in range(8):
    plt.scatter(x[y_kmeans == i, 0], x[y_kmeans == i, 1], s=100)
```
